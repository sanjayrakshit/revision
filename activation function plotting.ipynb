```python
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
lo, hi = -10, 10
sample_size = 1000
interval_size = (hi - lo) / sample_size
x_values = [lo+i*interval_size for i in range(sample_size)]
```

```python
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
y_value_tanh = np.tanh(x_values)
y_value_sigmoid = [sigmoid(i) for i in x_values]
```

```python
# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x_values, y_value_tanh, label="tanh(x)", color='blue')
plt.plot(x_values, y_value_sigmoid, label="sigmoid(x)", color='red')

# Adding title and labels
plt.title("tanh(x) vs sigmoid(x)")
plt.xlabel("x")
plt.ylabel("y")

# Adding a legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
def sigmoid_derivative(z):
    return sigmoid(z)*(1-sigmoid(z))

def tanh_derivative(z):
    return 1 - math.tanh(z)**2
```

```python
y_value_sigmoid_derv = [sigmoid_derivative(i) for i in x_values]
y_value_tanh_derv = [tanh_derivative(i) for i in x_values]
```

```python
# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x_values, y_value_tanh_derv, label="derv tanh(x)", color='blue')
plt.plot(x_values, y_value_sigmoid_derv, label="derv sigmoid(x)", color='red')

# Adding title and labels
plt.title("derivative of tanh(x) vs sigmoid(x)")
plt.xlabel("x")
plt.ylabel("y")

# Adding a legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
